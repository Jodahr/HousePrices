```python
lowCard = Pipeline([
    ('getLow', ppc.SplitCardinality(larger=False, threshold=5)),
    ('fill_nan', ppc.DFTransform(lambda X: X.fillna('NaN'))),
    ('oneHot', ppc.DFDummyTransformer())
])

highCard = Pipeline([
    ('getLarge', ppc.SplitCardinality(larger=True, threshold=5)),
    ('cardinality_scheme', ppc.HighCardinality(f=0.1, assumption='mean')),
    ('scaler', ppc.DFScaler(scaler=MinMaxScaler())),
])
```

```python
ppc.SplitCardinality(larger=False, threshold=19).fit_transform(X_train)
```

```python
cat_pipe = Pipeline([
    ('select_cat', ppc.DFTransform(lambda X: X.select_dtypes(include=[object]))),
    ('drop_nulls', ppc.DropTooManyNulls(threshold=0.3)),
    ('drop_unique', ppc.DropTooManyUnique(threshold=30)),
     ('par_pipe', ppc.DFFeatureUnion([
        ('highCard', highCard), 
         ('lowCard', lowCard),
#    ('fill_nan', ppc.DFTransform(lambda X: X.fillna('NaN'))),
#    ('oneHot', ppc.DFDummyTransformer())
    ]))
])
```

```python
num_pipe = Pipeline([
    ('select_num', ppc.DFTransform(lambda X: X.select_dtypes(exclude=[object]))),
    ('fill_nan', ppc.DFImputer(imputer=Imputer(strategy='median'))),
    ('scaler', ppc.DFScaler(scaler=MinMaxScaler())),
])
```

```python
pipe = Pipeline([
    ('dropCols', ppc.DropColumns(cols=['Id'])),
    ('parallel_pipe', ppc.DFFeatureUnion([
        ('cat',cat_pipe), ('num', num_pipe)
    ])),
    #('scaler', ppc.DFScaler(scaler=MinMaxScaler())),
    #('knn', KNeighborsRegressor())
    #('rf_reg', RandomForestRegressor(n_estimators=1000, n_jobs=10))
    ('rf_reg', MLPRegressor())
    
])
```
